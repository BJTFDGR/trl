```python
import numpy as np
import matplotlib.pyplot as plt

# Data
comparisons = ['P+T', 'P']
mo_values = [0.1405964759, 0.1436677667 ]
mp_values = [0.2677178825, 0.07725636903]

# mo_values = [0.1405964759, 0.07725636903, 0.2677178825]
# mp_values = [0.2677178825, 0.1436677667, 0.1436677667]


x = np.arange(len(comparisons))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, mo_values, width, label='Mo')
rects2 = ax.bar(x + width/2, mp_values, width, label='Mp')

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_xlabel('Comparisons')
ax.set_ylabel('Toxicity')
ax.set_title('Comparison of Mo and Mp')
ax.set_xticks(x)
ax.set_xticklabels(comparisons)
ax.legend()

# Attach a text label above each bar in *rects*, displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
